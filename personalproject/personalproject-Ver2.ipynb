```python
import pandas as pd
```

```python
location = "./usdapopulation.xls"
df = pd.read_excel(location)
df.head()
```

```python
df_county = df['Area_Name']
```

```python
df_county.head()
```

```python
import nltk 
from nltk.probability import FreqDist
```

```python
fd_county = FreqDist(df_county) #This function gives the frequency
```

```python
fd_county 
```

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(background_color = 'white').generate_from_frequencies(fd_county)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
