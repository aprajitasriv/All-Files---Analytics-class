```python
location = "./usdaunemployment.xls"
df_unemploy = pd.read_excel(location)
df_unemploy.head()
```

```python
df_Unemploy_by_yr = df_unemploy[['FIPStxt','State','Unemployment_rate_2013','Unemployment_rate_2014','Unemployment_rate_2015','Unemployment_rate_2016','Unemployment_rate_2017']]
df_Unemploy_by_yr
```

```python
#creat new database with 3 states
df_states_unemploy = df_Unemploy_by_yr.iloc[[1, 102,118]]
df_states_unemploy
```

```python
Col_Names=df_states_unemploy.iloc[:,2:]
# by putting a : at the begining, in [:] - it references the row
```

```python
Col_Str=' + '.join(Col_Names)
Col_Str
```

```python
df_states_unemploy['mean']=df_states_unemploy.eval(Col_Str)/len(df_states_unemploy.columns)
```
